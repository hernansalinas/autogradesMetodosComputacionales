#El método de Runge-Kutta de orden cuatro

El método de Runge-Kutta de orden cuatro es un método numérico para resolver ecuaciones diferenciales ordinarias. Se basa en aproximar la solución a

\begin{equation}
\frac{\mathrm{d}y}{\mathrm{d}t} = f(t, y)
\end{equation}

con condiciones iniciales $y(t_0)=y_0$ a:


\begin{equation}
y_{n+1} = y_n + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4)
\end{equation}

donde $h$ es el tamaño del paso y
\begin{equation}
k_1 = f(t_n, y_n)
\end{equation}

\begin{equation}
k_2 = f(t_n + \frac{h}{2}, y_n + \frac{h}{2}k_1)
\end{equation}

\begin{equation}
k_3 = f(t_n + \frac{h}{2}, y_n + \frac{h}{2}k_2)
\end{equation}

\begin{equation}
k_4 = f(t_n + h, y_n + hk_3)
\end{equation}

El orden de convergencia del método es cuatro, lo que significa que el error local es proporcional a $h^5$ y el error global es proporcional a $h^4$.

# El Sistema de Lorentz

## Introducción
El sistema de Lorentz es un conjunto de ecuaciones diferenciales ordinarias (EDOs) de primer orden que surgieron como una simplificación de un modelo meteorológico. Este sistema fue desarrollado por el meteorólogo y matemático Edward Lorentz en 1963 mientras estudiaba el comportamiento atmosférico. Las ecuaciones de Lorentz son famosas por su papel en la teoría del caos, mostrando cómo sistemas deterministas pueden comportarse de manera impredecible.

## Ecuaciones del Sistema de Lorentz
El sistema de Lorentz se define mediante las siguientes tres ecuaciones diferenciales no lineales:



$$
\begin{aligned}
\frac{dx}{dt} &= \sigma (y - x) \\
\frac{dy}{dt} &= x (\rho - z) - y \\
\frac{dz}{dt} &= xy - \beta z
\end{aligned}
$$

donde $x$,$y$, y $z$ representan las variables del sistema, y $\sigma$, $\rho$, y $\beta$ son parámetros constantes.

### Parámetros
- $\sigma$ (sigma): Es el número de Prandtl, que representa la relación entre la viscosidad del fluido y la difusión térmica.
- $\rho$ (rho): Es el número de Rayleigh, que se relaciona con la diferencia de temperatura entre el techo y el suelo del sistema.
- $\beta$ (beta): Está relacionado con la geometría del sistema y usualmente se define como 8/3 en los estudios de Lorentz.

## Comportamiento y Caos
El sistema de Lorentz es un ejemplo clásico de un sistema dinámico caótico. A pesar de ser determinista, lo que significa que sus futuras trayectorias están completamente determinadas por sus condiciones iniciales, el sistema muestra una gran sensibilidad a estas condiciones iniciales. Esto implica que dos conjuntos de condiciones iniciales que son arbitrariamente cercanas entre sí pueden conducir a comportamientos a largo plazo completamente diferentes, una característica conocida como el "efecto mariposa".

## Atractor de Lorentz
Una de las características más notables del sistema de Lorentz es el "atrayente de Lorentz", una estructura fractal y caótica que describe cómo las trayectorias de las soluciones se acercan a un conjunto limitado en el espacio de fases. Este atrayente es un ejemplo clásico de un sistema caótico y ha sido ampliamente estudiado en la teoría del caos y la dinámica no lineal.

## Aplicaciones
El sistema de Lorentz no solo es importante en la meteorología, sino que también ha encontrado aplicaciones en otros campos como la ingeniería, la física y la biología. El estudio de estos sistemas caóticos ha permitido una mejor comprensión de fenómenos complejos y la naturaleza del caos en diversos sistemas.
